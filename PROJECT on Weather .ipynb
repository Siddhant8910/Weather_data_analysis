```python
# Extract relevant columns in order to get the required stats using numpys array
# Extract relevant numeric columns as NumPy arrays
temperature = data['temperature (c)'].to_numpy()
apparent_temperature = data['apparent temperature (c)'].to_numpy()
humidity = data['humidity'].to_numpy()
wind_speed = data['wind speed (km/h)'].to_numpy()
wind_bearing = data['wind bearing (degrees)'].to_numpy()
visibility = data['visibility (km)'].to_numpy()
pressure = data['pressure (millibars)'].to_numpy()

```

time series analysis

```python
# Convert to datetime with UTC
data['formatted date'] = pd.to_datetime(data['formatted date'], utc=True)

# Optionally remove timezone information
data['formatted date'] = data['formatted date'].dt.tz_localize(None)

# Extract additional features
data['year'] = data['formatted date'].dt.year
data['month'] = data['formatted date'].dt.month
data['day'] = data['formatted date'].dt.day

# Display the updated DataFrame
print(data.head())
```

```python
# Select only numeric columns
numeric_columns = data.select_dtypes(include='number').columns

# Include 'formatted date' in the DataFrame to maintain datetime index for resampling
numeric_data = data[numeric_columns].copy()
numeric_data['formatted date'] = data['formatted date']

# Set the 'formatted date' as the index
numeric_data.set_index('formatted date', inplace=True)

# Resample data to monthly averages
monthly_data = numeric_data.resample('M').mean()

# Display the resampled DataFrame
monthly_data
```

From the above result we can find the monthly mean of a neumeric values from the numeric columns


```python
import matplotlib.pyplot as plt

# Plot temperature trend
plt.figure(figsize=(15, 8))
plt.plot(monthly_data.index, monthly_data['temperature (c)'], label='Temperature (°C)', marker='o')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Change Over a Monthly Period')
plt.legend()

# Find and annotate peak value
peak_temp = monthly_data['temperature (c)'].max()
peak_date = monthly_data['temperature (c)'].idxmax()

# Print the exact date and temperature of the peak value
print(f"Peak temperature: {peak_temp:.2f}°C on {peak_date.strftime('%Y-%m-%d')}")

plt.annotate(f'Peak: {peak_temp:.2f}°C', 
             xy=(peak_date, peak_temp), 
             xytext=(peak_date, peak_temp + 3),
             arrowprops=dict(facecolor='black', shrink=0.005),
             ha='center')

# Find and annotate minimum value
min_temp = monthly_data['temperature (c)'].min()
min_date = monthly_data['temperature (c)'].idxmin()

# Print the exact date and temperature of the minimum value
print(f"Minimum temperature: {min_temp:.2f}°C on {min_date.strftime('%Y-%m-%d')}")

plt.annotate(f'Min: {min_temp:.2f}°C', 
             xy=(min_date, min_temp), 
             xytext=(min_date, min_temp - 2),  # Adjust the offset to avoid overlap
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

plt.show()
```

From the above plot we can observed that the minimum and maximum temprature that is  24.56 and -5.14 degree celcious was

```python
# Calculate mean and standard deviation for the numeric columns
numeric_columns = ['temperature (c)', 'apparent temperature (c)', 'humidity', 'wind speed (km/h)', 
                   'wind bearing (degrees)', 'visibility (km)', 'pressure (millibars)']

mean_std = data[numeric_columns].agg(['mean', 'std'])

# Define thresholds for anomalies (e.g., mean ± 2*std)
thresholds = {}
for column in numeric_columns:
    thresholds[column] = {
        'high': mean_std.loc['mean', column] + 2 * mean_std.loc['std', column],
        'low': mean_std.loc['mean', column] - 2 * mean_std.loc['std', column]
    }

# Identify anomalies
anomalies = pd.DataFrame()
for column in numeric_columns:
    high_anomalies = data[data[column] > thresholds[column]['high']]
    low_anomalies = data[data[column] < thresholds[column]['low']]
    anomalies = pd.concat([anomalies, high_anomalies, low_anomalies])

# Remove duplicates (in case of multiple columns having the same anomaly)
anomalies = anomalies.drop_duplicates()
print("Detected Anomalies:")
print(anomalies[['formatted date', 'summary', 'temperature (c)', 'apparent temperature (c)', 'humidity', 
                 'wind speed (km/h)', 'wind bearing (degrees)', 'visibility (km)', 'pressure (millibars)']])
```

Temperature:

There are significant anomalies in temperature readings, such as the high values for the "Mostly Cloudy" condition on August 1, 2006, and consistent low values on March 23, 2016, despite the clear condition.

Apparent Temperature:

Apparent temperatures closely follow actual temperatures but reflect the same anomalies, indicating discrepancies in the data.
Humidity:

Humidity levels show irregularities, with extremely low values on August 1, 2006, during "Mostly Cloudy" conditions, and high values on March 23, 2016, despite the clear condition.
